```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
worldDF = spark.read.option('inferSchema','true').csv('datasets/world.txt')
worldDF = worldDF.withColumnRenamed('_c0','Country')
worldDF = worldDF.withColumnRenamed('_c1','Continent')
worldDF = worldDF.withColumnRenamed('_c2','CapitalCity')
worldDF = worldDF.withColumnRenamed('_c3','Population')
worldDF = worldDF.drop('Continent')
worldDF = worldDF.sort('Population',ascending=False)
worldDF = worldDF.filter(worldDF['Population']>1000000000)
worldDF.show()
worldDF.printSchema()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
worldDF = spark.read.csv('datasets/world.txt')
worldDF = worldDF.withColumnRenamed('_c0','Country')
worldDF = worldDF.withColumnRenamed('_c1','Continent')
worldDF = worldDF.withColumnRenamed('_c2','CapitalCity')
worldDF = worldDF.withColumnRenamed('_c3','Population')

worldDF = worldDF.withColumn('Population2',worldDF['Population'].cast('float'))
worldDF.printSchema()
worldDF.show()
```
