```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
eqDF = spark.read.option('header','true').option('delimiter','\t').option('inferSchema','true').csv('datasets/EarthquakeDataset-latest.txt')
eqDF = eqDF.select('Date','Origin Time','Latitude','Longitude','xM','Location')
eqDF = eqDF.withColumn('NewColumn', eqDF['Latitude'] + 2*eqDF['Longitude'])
eqDF.show()
eqDF.printSchema()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
actorsDF = spark.read.option('header','true').option('inferSchema','true').csv('datasets/join-actors.txt')
seriesDF = spark.read.option('header','true').option('inferSchema','true').csv('datasets/join-series.txt')

print('actor count:' ,actorsDF.count())
actorsDF.show()
seriesDF.show()
joinedDF = actorsDF.join(seriesDF,'movieid','inner') #all sql joins will work here (inner,outer,left,right)
joinedDF.show(50)
```
